# Data Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title ('Reservation status count')
plt.bar(['Not cancelled' , 'Cancelled'],df['is_canceled'].value_counts(), edgecolor = 'k' , width = 0.7)
plt.show()
```

```python
# import matplotlib.pyplot as plt

cancelled_counts = df['is_canceled'].value_counts()

# Create a bar plot
plt.figure(figsize=(4, 5))  
ax = cancelled_counts.plot(kind='bar')
plt.title('Cancellation Counts by Category')
plt.xlabel('Cancellation Status')
plt.ylabel('Count')
plt.xticks(rotation=0) 
# plt.yticks(rotation=90) 

# Adding labels to the bars
for index, value in enumerate(cancelled_counts):
    plt.text(index, value, str(value), ha='center', va='bottom')

# Adding labels for 'cancelled' and 'not cancelled'
ax.set_xticklabels(['Not Cancelled', 'Cancelled'])

plt.show()  # Display the plot

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Get the handles and labels for the legend
legend_labels, _ = ax1.get_legend_handles_labels()

# Create custom legend handles with labels and matching colors
custom_legend = [
    plt.Line2D([0], [0], color='blue', lw=3),
    plt.Line2D([0], [0], color='orange', lw=3)
]
legend_labels = ['Not Canceled', 'Canceled']

# Create the legend with custom handles and labels
ax1.legend(custom_legend, legend_labels, title="Cancellation Status", loc="upper right")

plt.title("Reservation status in different hotels", size=20)
plt.xlabel("Hotel")
plt.ylabel("Number of Reservations")
plt.show()
```

```python
import pandas as pd

# Filter for City Hotel
city_hotel = df[df['hotel'] == "City Hotel"]

# Calculate the average daily rate for City Hotel
average_daily_rate_city = city_hotel['adr'].mean()

# Filter for Resort Hotel
resort_hotel = df[df['hotel'] == "Resort Hotel"]

# Calculate the average daily rate for Resort Hotel
average_daily_rate_resort = resort_hotel['adr'].mean()

print("Average Daily Rate for City Hotel:", average_daily_rate_city)
print("Average Daily Rate for Resort Hotel:", average_daily_rate_resort)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.title("Average Daily Rate in City and Resort Hotel", fontsize=30)

# Assuming you have defined 'resort_hotel' and 'city_hotel' DataFrames
plt.plot(resort_hotel.index, resort_hotel['adr'], label="Resort Hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label="City Hotel")

plt.legend(fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have defined 'resort_hotel' and 'city_hotel' DataFrames
plt.figure(figsize=(12, 6))
plt.title("Average Daily Rate by Day", fontsize=16)

# Create a scatter plot for Resort Hotel
plt.scatter(resort_hotel.index, resort_hotel['adr'], label="Resort Hotel", color="royalblue", s=10, alpha=0.7)

# Create a scatter plot for City Hotel
plt.scatter(city_hotel.index, city_hotel['adr'], label="City Hotel", color="coral", s=10, alpha=0.7)

# Customize the tick label font size
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add labels to the axes
plt.xlabel("Day", fontsize=14)
plt.ylabel("Average Daily Rate", fontsize=14)

# Add a legend
plt.legend(fontsize=12)

# Show the plot
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
